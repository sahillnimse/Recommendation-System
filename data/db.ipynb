```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('user_clusters.db')

# Query all data from the recommendations table
df_db_view = pd.read_sql_query("SELECT * FROM user_clusters", conn)
conn.close()

# Display the results
print(f"Database contains {len(df_db_view)} records.")
display(df_db_view.head(20))
```

```python
import sqlite3

old = sqlite3.connect('user_clusters.db')
new = sqlite3.connect('user_clusters_new.db')

# Step 1: Add Name column to old DB
old.execute("ALTER TABLE user_clusters ADD COLUMN Name TEXT")

# Step 2: Copy names from new DB
new_names = new.execute("SELECT user_id, Name FROM user_clusters").fetchall()

# Step 3: Update old DB
old.executemany("UPDATE user_clusters SET Name=? WHERE user_id=?", 
                [(name, uid) for uid, name in new_names])

old.commit()
old.close()
new.close()
print("Done!")
```
